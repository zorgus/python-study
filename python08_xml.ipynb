```python
import xml.etree.ElementTree as ET

tree = ET.parse("data/song.xml")
rss = tree.getroot()
#for songElm in rss.findall('./channel/song'):
for songElm in rss.findall('.//song'):
    print(songElm.attrib['sname'])
    print(songElm.findtext('title'))
    print(songElm.findtext('singer'))
```

```python
import xml.etree.ElementTree as ET
import urllib.request as REQ

kurl = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'
response = REQ.urlopen(kurl)
data = response.read()
data = data.decode('utf-8')
rss = ET.fromstring(data)
for loc in rss.findall('.//location'):
    print(loc.findtext('city'))
    print("="*30)
    for data in loc.findall('.//data'):
        print("날짜: {}".format(data.findtext('tmEf')))
        print("날씨: {}".format(data.findtext('wf')))
        print("최저: {}".format(data.findtext('tmn')))
        print("최고: {}".format(data.findtext('tmx')))
        print("-"*30)
```

```python
### Beautifulsoup (3rd party library) from web
from bs4 import BeautifulSoup
import urllib.request as REQ

jurl = 'http://rss.joins.com/joins_news_list.xml'
response = REQ.urlopen(jurl)
soup = BeautifulSoup(response, 'html.parser')
for item in soup.findAll('item'):
    print("-"*80)
    print("제목: {}".format(item.title.string))
    print("내용: {}".format(item.description.string))
```

```python
### element 규칙
# child dot(.), parent
# findNext, findPrevious
# 
from bs4 import BeautifulSoup

fp = open('song.xml')
soup = BeautifulSoup(fp, 'html.parser')
fp.close()
songs = soup.findAll('song')
print(songs)
print(songs[0])
print("-"*80)

song = soup.find('song')
print(song.title)
print(song.singer)
# print(song.parent)
print(song.findNext('song').title)
print(song.findNext('song').singer)
print(song.findNext('song').findNext('song').title)
print(song.findNext('song').findNext('song').singer)
print("-"*80)

song1 = song.findNext('song')
print(song1.findPrevious('song'))
```

```python
### Beautifulsoup (3rd party library)
# element 추가(insert)

from bs4 import BeautifulSoup

fp = open('data/song.xml')
soup = BeautifulSoup(fp, 'html.parser')
channel = soup.find('channel')
song = soup.new_tag('song', sname='sname4')
title = soup.new_tag('title')
title.string = 'title4'
singer = soup.new_tag('singer')
singer.string = 'singer4'
song.append(title)
song.append(singer)
channel.insert(2, song)

print(soup.prettify())
```
