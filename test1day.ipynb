```python
### 1일차 과제
'''
주소록을 입력받아 결과를 출력하시요
(함수로 구성 만드시요)

메인 메뉴)
     1.입력
     2.출력
     3.검색
     4.종료
     번호를 입력하세요:

입력)

1. 입력

이름:
나이:
주소:
계속 입력하시겠습니까(y/n)? 

입력이 종료되면 메인메뉴로


2.출력
--------------------------------
    이름        나이        주소
-------------------------------
     ㅁㅁㅁ      20        서울시
     ㅁㅁㅁ   

3.검색
검색할 이름을 입력하세요:
--------------------------------
    이름        나이        주소
-------------------------------
     ㅁㅁㅁ      20        서울시
'''
```

```python
isEnd = False
Contact = []

def MainMenu():
    print("1.입력")
    print("2.출력")
    print("3.검색")
    print("4.종료")
    num = input("번호를 입력하세요:")
    if not num.isdigit():
        return 0
    return int(num)

def InputContact():
    continueInput = 'y'
    while(continueInput == 'y'):
        name = input("이름:")
        age = input("나이:")
        addr = input("주소:")
        item = {'name':name, 'age':age, 'addr':addr}
        Contact.append(item)
        continueInput = input("계속 입력하시겠습니까(y/n)? ")

def ShowContact():
    print("--------------------------------")
    print("%10s%10s%10s" %("이름","나이","주소"))
    print("-------------------------------")
    for c in Contact:
        print("%10s%10s%10s" % (c['name'],c['age'],c['addr']))

def SearchContact():
    name = input("검색할 이름을 입력하세요:")
    print("--------------------------------")
    print("%8s%8s%8s" %("이름","나이","주소"))
    print("-------------------------------")
    for c in Contact:
        if c['name'] == name:
          print("%8s%8s%8s" % (c['name'],c['age'],c['addr']))


while (not isEnd):
    choice = MainMenu()
    if choice == 1:
        InputContact()
    elif choice == 2:
        ShowContact()
    elif choice == 3:
        SearchContact()
    elif choice == 4:
        isEnd = True

```
