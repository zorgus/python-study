```python
'''
	서울 경기 날씨
1. 보기
2. 최고기온
3. 지역별 최저기온
4. 평균기온
5. 비가오는 지역과 날짜
6. 검색지역:
7. 엑셀저장(차트포함)

2번양식

X 지역 최고기온은 x 월 x 일 xx 시 x도입니다

3번양식
서울: 최저기온은 x 월 x 일 xx 시 x도입니다
인천: 최저기온은 x 월 x 일 xx 시 x도입니다
...

4번양식
서울경기 최고기온 평균기온은 x 도 입니다.
서울경기 최저기온 평균기온은 x 도 입니다.

5번양식
예)서울: 2017-10-15  00:00
xx:xx-xx-xx xx:xx
.....

6. 검색지역입력(서울,수원,파주..):
날짜:
날씨:
최저:
최고:
....

7. 엑셀로저장하시겠습니까(y/n)?

'''
```

```python
from bs4 import BeautifulSoup
from statistics import mean
import urllib.request as REQ

class Weather:
    url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'
    def __init__(self, loc):
        self.location = loc
        self.data = []
        self.get()
   
    def get(self):
        response = REQ.urlopen(Weather.url)
        soup = BeautifulSoup(response, 'html.parser')
        for loc in soup.findAll('location'):
            if self.location == '서울' and loc.city.string == '서울':
                self.data = loc.findAll('data')
            else:
                self.data.extend(loc.findAll('data'))
                
    def show(self):
        for data in self.data:
            print("날짜: {}".format(data.tmef.string))
            print("날씨: {}".format(data.wf.string))
            print("최저: {}".format(data.tmn.string))
            print("최고: {}".format(data.tmx.string))
            print("-"*30)
    def gethigh(self):
        htempdata = None
        for data in self.data:
            if htempdata == None or int(htempdata.tmx.string) < int(data.tmx.string):
                htempdata = data
        return (htempdata.tmef.string, htempdata.tmx.string)
                    
    def getlow(self):
        mtempdata = None
        for data in self.data:
            if mtempdata == None or int(mtempdata.tmn.string) > int(data.tmn.string):
                mtempdata = data
        return (mtempdata.tmef.string, mtempdata.tmn.string)
    def getmeanhigh(self):
        return mean([int(n.tmx.string) for n in self.data])
    def getmeanlow(self):
        return mean([int(n.tmn.string) for n in self.data])
    
```

```python
class Menu:
    def __init__(self):
        self.data = {}
        self.data['서울'] = Weather('서울')
        self.data['경기'] = Weather('경기')
    def main(self):
        menu = {'1':self.showAll, '2':self.showHighTemp,
                '3':self.showLowTemp, '4':self.showMeanTemp,
                '5':self.searchLoc,
               }
        
        menulist = ("1. 보기",
                    "2. 최고기온",
                    "3. 지역별 최저기온",
                    "4. 평균기온",
                    "5. 검색지역")
        print("서울 경기 날씨")
        print("{}\n{}\n{}\n{}\n{}".format(*menulist), sep='\n')
        num = input("번호를 입력하세요:")
        menu.get(num, self.wrongInput)()

    def wrongInput(self):
        print("잘못입력했습니다.")

    def showAll(self):
        for k,v in self.data.items():
            print(k)
            print("="*30)
            v.show()
    def showHighTemp(self):
        for k,v in self.data.items():
            date, temp = v.gethigh()
            print("{} 지역 최고기온은 {} {}도입니다".format(k, date, temp))
    def showLowTemp(self):
        for k,v in self.data.items():
            date, temp = v.getlow()
            print("{} 지역 최저기온은 {} {}도입니다".format(k, date, temp))
    def showMeanTemp(self):
        meanlist = [0,0]
        for k,v in self.data.items():
            meanlist[0] += v.getmeanhigh()
            meanlist[1] += v.getmeanlow()
        print("서울경기 최고기온 평균기온은 {:.2f} 도 입니다.".format(meanlist[0]/ len(self.data)))
        print("서울경기 최저기온 평균기온은 {:.2f} 도 입니다.".format(meanlist[1]/ len(self.data)))
    def searchLoc(self):
        loc = input("검색지역입력:")
        self.data.get(loc, lambda :print("잘못된 지역입니다")).show()
     


```

```python
menu = Menu()
menu.main()
```

```python
from bs4 import BeautifulSoup
from statistics import mean
import urllib.request as REQ

class Weather2:
    url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'
    def __init__(self):
        self.data = {}
        self.get()
   
    def get(self):
        response = REQ.urlopen(Weather2.url)
        soup = BeautifulSoup(response, 'html.parser')
        for loc in soup.findAll('location'):
            citystring = loc.city.string
            self.data[citystring] = []
            for data in loc.findAll('data'):
                self.data[citystring].append(
                            {'city': citystring,
                            'tmef': data.tmef.string,
                            'wf': data.wf.string,
                            'tmn': int(data.tmn.string),
                            'tmx': int(data.tmx.string)})
                
    def show(self, city='all'):
        for data in self.getcitydata(city):
            print("날짜: {}".format(data['tmef']))
            print("날씨: {}".format(data['wf']))
            print("최저: {}".format(data['tmn']))
            print("최고: {}".format(data['tmx']))
            print("-"*30)

    def getcitylist(self):
        return self.data.keys()

    def getcitydata(self, city = 'all'):
        if city == 'all':
            ### 이중 for loop를 이용하여 list 생성
            return [m for n in self.data.values() for m in n]
        else:
            return self.data[city]

    def gethigh(self, city='all'):
        return max(self.getcitydata(city), key=lambda k:k['tmx'])
    def getlow(self, city='all'):
        return min(self.getcitydata(city), key=lambda k:k['tmn'])
    def getmeanhigh(self, city='all'):
        return mean([n['tmx'] for n in self.getcitydata(city)])
    def getmeanlow(self, city='all'):
        return mean([n['tmn'] for n in self.getcitydata(city)])

if __name__ == '__main__':
    w = Weather2()
    print(w.gethigh())
    print(w.getlow())
    print(w.getmeanhigh())
    print(w.getmeanlow())
```
