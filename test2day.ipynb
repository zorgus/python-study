```python
'''
성적을 입력받아 결과를 출력하시요
(함수로 구성하고 별도 모듈파일을 만드시요)

메인 메뉴)
     1.입력
     2.출력
     3.검색
     4.종료
     번호를 입력하세요:

입력)

1. 입력

이름:
국어:
영어:
수학:
계속 입력하시겠습니까(y/n)? 
입력이 종료되면 메인메뉴로

2.출력
-----------------------------------------------------------------
    이름        국어        영어	수학	총점	평균
----------------------------------------------------------------
     ㅁㅁㅁ      20        30	40
	...
----------------------------------------------------------------
총점: 국어 xx  영어 xx 수학
평균: 국어 xx  영어 xx 수학
최고점수: 국어 xx  영어 xx 수학

3.검색
검색할 이름을 입력하세요:
----------------------------------------
    이름        국어        영어	수학
----------------------------------------
'''
```

```python
# coding: utf-8

class Scores:
    def __init__(self):
        self.data = []
    def append(self, name, kor, eng, math):
        self.data.append({'name':name, 'kor':kor, 'eng':eng, 'math':math})
    def iterate(self):
        for data in self.data:
            yield data
    def search(self, value, key='name'):
        for d in self.iterate():
            if d[key] == value:
                yield d
    def max(self, key='kor'):
        maxi = max(self.data, key=lambda k:k[key])
        return maxi[key]
    def mean(self, key='kor'):
        #return mean([n[key] for n in self.data])
        if (len(self.data) > 0):
            return self.sum(key)/len(self.data)
        else:
            return 0
    def sum(self, key='kor'):
        return sum([n[key] for n in self.data])
```

```python
# coding: utf-8

#from score import Scores

class Menu:
    def __init__(self):
        self.scores = Scores()
        self.isexit = False
    def loop(self):
        menu = {'1':self.inputData, '2':self.outputData,
                '3':self.searchData, '4':self.exit }
        while not self.isexit:
            print("1.입력", "2.출력", "3.검색", "4.종료", sep='\n')
            num = input("번호를 입력하세요:")
            menu.get(num, self.wrongInput)()

    def wrongInput(self):
        print("잘못입력했습니다.")

    def showTitle(self):
        items = ('이름', '국어', '영어', '수학', '총점', '평균')
        print("-"*70)
        print("{:10}{:10}{:10}{:10}{:10}{:10}".format(*items))
        print("-"*70)

    def showOne(self, score):
        scoreonly = [n for n in score.values() if type(n) == int ]
        sumi = sum(scoreonly)
        meani = sum(scoreonly) / len(scoreonly)
        print("{name:10}{kor:10}{eng:10}{math:10}{:10}{:10.2f}".format(sumi, meani, **score))

    def exit(self):
        self.isexit = True

    def inputData(self):
        while True:
            name = input("이름:")
            kor = input("국어:")
            eng = input("영어:")
            math = input("수학:")
            self.scores.append(name, int(kor), int(eng), int(math))
            yn = input("계속 입력하시겠습니까(y/n)? ")
            if yn == 'n':
                break

    def outputData(self):
        self.showTitle()
        for s in self.scores.iterate():
            self.showOne(s)
        print("총점: 국어 {}  영어 {}  수학 {}".format(
            self.scores.sum(key='kor'),
            self.scores.sum(key='eng'),
            self.scores.sum(key='math')))
        print("평균: 국어 {:.2f}  영어 {:.2f}  수학 {:.2f}".format(
            self.scores.mean(key='kor'),
            self.scores.mean(key='eng'),
            self.scores.mean(key='math')))
        print("최고점수: 국어 {}  영어 {}  수학 {}".format(
            self.scores.max(key='kor'),
            self.scores.max(key='eng'),
            self.scores.max(key='math')))

    def searchData(self):
        name = input("검색할 이름을 입력하세요:")
        s = self.scores.search(name)
        self.showTitle()
        for n in s:
            self.showOne(n)
```

```python
#from menu import Menu

mainMenu = Menu()
mainMenu.loop()
```
